```python
#Load the Training Data from Excel to dataframe - train_data
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

```python
#Load the Test Data from Excel to dataframe - test_data
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.head()
```

```python
#How many women survived from Training Data? 
women = train_data.loc[ train_data.Sex == 'female' ][ 'Survived' ]
women_surv_rate = sum( women )/len( women )
print( "The % of Women  who survived is: ", women_surv_rate )
```

```python
#How many men survived from the Training Data?
men = train_data.loc[ train_data.Sex == 'male' ][ 'Survived' ]
men_surv_rate = sum( men )/len( men )
print( "The % of Men who survived is: ", men_surv_rate )
```

```python
#Use Machine Learning, Supervised Learning Method - Random Forest.
from sklearn.ensemble import RandomForestClassifier

y = train_data[ 'Survived' ]

feat = [ 'Pclass', 'Sex', 'SibSp', 'Parch' ] #Look for patterns in the columns - Interested Features.

X_train = pd.get_dummies( train_data[ feat ] ) #Training input to the model.

X_test = pd.get_dummies( test_data[ feat ] ) #Testing input to the model.

titanic_model = RandomForestClassifier( n_estimators = 100, max_depth = 5, random_state = 1 )

titanic_model.fit( X_train, y )
```

```python
#Output this into an Excel file as mentioned in the Question:
predicted = titanic_model.predict( X_test )
output = pd.DataFrame( { 'PassengerId': test_data.PassengerId, 'Survived': predicted } )

#Writing into CSV
output.to_csv( 'Submission.csv', index = False )

print( "My Submission is saved Successfully" )
```
